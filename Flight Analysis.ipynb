```python
# Loading libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
wa =sns.load_dataset('flights')
```

```python
wa.info()
```

```python
# The graphical representation of passengers over the years
plt.figure(figsize=(10,5))
sns.barplot(x='year', y='passengers', data=wa, errorbar=None)
plt.title('Total Passengers per Year')
plt.show()
```

```python
### The flow of passengers increased yearly.
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='month', y='passengers', data=wa, errorbar=None)
plt.title('Total Passengers travelling per Month')
plt.show()
```

```python
### July and August had the highest number of passengers.
```

```python
# Data transformation
wa1=pd.melt(wa, id_vars=['month', 'passengers'], value_vars=['year'])
wa1.head()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='month', y='passengers', hue= 'value', data=wa1, errorbar=None)
plt.legend(title ='value', bbox_to_anchor =(0.3,-0.1), ncol=5)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='value', y='passengers', hue= 'month', data=wa1, errorbar=None)
plt.legend(title ='value', bbox_to_anchor =(0.3,-0.1), ncol=5)
plt.show()
```

```python
wa2=wa.sort_values(by=['passengers'], ascending=False)
wa2
```

```python
wa2.describe()
```

```python
### Analysing the data when passengers are above the mean value and 75%.
```

```python
# The mean(280) is used to analyse
wa2a=wa2[wa2.passengers>280]
wa2a
```

```python
wa2a=wa2a.sort_values(by=['month'], ascending=False)
wa2a
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='month', y='passengers',  data=wa2a)
plt.title('The monthly analysis over average')
plt.show()
```

```python
### There is an expecatancy of more passengers in July and August
```

```python
wa2b=wa2[wa2.passengers<280]
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='month', y='passengers',  data=wa2b)
plt.title('The monthly analysis below average')
plt.show()
```
