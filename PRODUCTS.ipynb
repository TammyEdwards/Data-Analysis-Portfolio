```python
# Loading libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

#### 1.3 Cleaning Data

```python
# Renumbering 
fd.index =np.arange(1, len(fd) +1)
```

```python
# Adding column name
fd.index.names =['S/N']
```

```python
for col in fd.columns:
    print(col)
```

```python
fd.head()
```

```python
# Dropping columns
fd=fd.loc[:, fd.columns.drop(['Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12','Unnamed: 13'])]
fd.head()
```

```python
# Filtering data
fd1 = fd[fd['Total '] > 0]
fd1.head()
```

```python
# Arranging data in ascending order
fd1 = fd.sort_values(by =['Total '], ascending =False)
fd1.head()
```

```python
# Dropping columns 
fd2a=fd1.loc[:, fd1.columns.drop(['Regions ','Total ', 'week 1', 'week 2','week 3', 'week 4'])]
fd2a.head()
```

```python
# Dropping columns 
fd2b=fd1.loc[:, fd1.columns.drop(['Regions ','Total ', 'Product A', 'Product B', 'Product C'])]
fd2b.head()
```

```python
#Rearranging data 
fd3= pd.melt(fd1, id_vars=['Regions ', 'Total ', 'Zones '], value_vars=['Product A', 'Product B', 'Product C'])
fd3.head()
```

```python
#Rearranging data 
fd3a= pd.melt(fd1, id_vars=['week 1', 'week 2','week 3', 'week 4'], value_vars=['Product A', 'Product B', 'Product C'])
fd3a.head()
```

```python
#Rearranging data 
fd4=pd.melt(fd1, id_vars=['Regions ', 'Total '], value_vars=[ 'week 1', 'week 2','week 3', 'week 4'])
fd4.head()
```

```python
#Rearranging data 
fd5=pd.melt(fd1, id_vars=['Zones ', 'Total '], value_vars=[ 'week 1', 'week 2','week 3', 'week 4'])
fd5.head()
```

#### 1.4 Data Interpretation and Visualization

### i. Regional performance

```python
# Plotting a bar chart
plt.figure(figsize=(20,5))
ax=sns.swarmplot(x='Regions ', y=  'Total ' , data=fd1)
for i in ax.containers:
    ax.bar_label(i,)
plt.xticks(rotation = 90)
plt.title('Regional performance')
plt.show()
```

```python
# Plotting a bar chart
plt.figure(figsize=(14,5))
ax=sns.barplot(x=  'Total ', y= 'Regions ', data=fd1.head())
for i in ax.containers:
    ax.bar_label(i,)
plt.title('Top performing regions')
sns.set_palette('tab10')
plt.show()
```

```python
sns.barplot(x='Regions ', y ='value' ,  data=fd4, hue='variable', errorbar=None)
plt.xticks(rotation = 90)
plt.grid()
sns.set(rc={'figure.figsize':(20,5)})
sns.dark_palette('red')
plt.title('Regional weekly performance')
plt.show()
```

```python
sns.barplot(x='Regions ', y ='value' ,  data=fd3, hue='variable', errorbar=None)
plt.xticks(rotation = 90)
plt.grid()
sns.set(rc={'figure.figsize':(20,5)})
plt.title('Regional product performance')
plt.show()
```

```python
# Plotting a distribution plot
plt.figure(figsize=(17,5))
sns.kdeplot(fd1)
plt.show()
```

### ii. Zonal performance
