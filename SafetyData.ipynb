```python
# Loading libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1.3 Cleaning Data

```python
# Droping null row
rep1=rep.drop([37])
```

```python
rep1.head()
```

```python
# Renumbering 
rep1.index =np.arange(1, len(rep1) +1)
```

```python
# Adding column name
rep1.index.names =['S/N']
```

```python
rep1.head()
```

```python
# Filtering data
rep2=rep1[rep1.Total> 0]
rep2.head()
```

```python
# Dropping columns
rep4=rep2.loc[:, rep2.columns.drop(['Insurgency', 'Accidents ', 'Unrelated to insurgency', 'Regions ','Total ', 'Zones '])]
rep4.head()
```

```python
rep8=pd.melt(rep4, id_vars=['Regions2', 'Total'], value_vars=[ 'week 1', 'week 2','week 3', 'week 4', 'week 5'])
rep8.head()
```

```python
# Dropping columns 
rep6=rep2.loc[:, rep2.columns.drop(['Regions ','Total ', 'week 1', 'week 2','week 3', 'week 4', 'week 5'])]
rep6.head()
```

```python
#Rearranging data 
rep7= pd.melt(rep6, id_vars=['Regions2', 'Total', 'Zones '], value_vars=['Insurgency', 'Accidents ', 'Unrelated to insurgency'])
rep7.head()
```

#### 1.4 Data Interpretation and Visualization

### i. Number of fatality in each state based on ALL causes

```python
# Making a smaller Dataframe
sa=rep2[['Regions2', 'Total']]
sa=sa.sort_values(by=['Total'],ascending=False)
```

```python
# Plotting a bar chart
plt.figure(figsize=(14,4))
ax=sns.barplot(x='Regions2', y=  'Total' , data=sa)
for i in ax.containers:
    ax.bar_label(i,)
plt.xticks(rotation = 90)
plt.title('Fatality of all states')
plt.show()
```

```python
# Plotting a bar chart
plt.figure(figsize=(14,4))
ax=sns.barplot(x='Regions2', y=  'Total' , data=sa.head())
for i in ax.containers:
    ax.bar_label(i,)
plt.title('Top 5 states')
plt.show()
```

```python
# Plotting a distribution plot
sns.displot(sa)
plt.show()
```

Interpretation: 
Over 15 states recorded deaths less than 20 and the 3 states recorded over 100 fatalities. 

### ii. Number of fatality in the Zones based on ALL causes

```python
# Making a smaller Dataframe
za=rep2[['Zones ', 'Total']]
za=za.sort_values(by=['Total'],ascending=False)
```

```python
# Plotting a box chart
plt.figure(figsize=(14,4))
sns.boxplot(x='Zones ', y=  'Total' , data=za)
plt.title('Fatality recorded for Zones')
plt.show()
```

```python
g=sns.FacetGrid(rep2, col='Zones ')
g.map(sns.barplot, 'Total', order=True, errorbar=None)
plt.show()
```

```python
# Plotting a pie chart
za['Zones '].value_counts().plot(kind='pie', figsize=(7,4), autopct='%0.0f%%', colors =['r', 'g', 'b', 'm', 'y', 'c'], startangle= -45, wedgeprops=dict(width=0.6))
plt.legend(title ='Zones', bbox_to_anchor =(2.3,0.5), ncol=2)
plt.title('Fatality recorded for Zones presented in Pie Chart')
plt.show()
```

Interpretation: 
The Northern region recorded more fatality than the Southern which avearged at 13%. 

### iii. weekly representation of fatality rate

```python
# Making a smaller Dataframe
wa=rep2[['week 1', 'week 2', 'week 3', 'week 4', 'week 5']]
wa=wa.sort_values(by=['week 1'],ascending=False)
```

```python
# Plotting a bar chart
plt.figure(figsize=(10, 4))
sns.barplot(data=wa, errorbar=None)
plt.title('Wwekly representation of fatality rate')
plt.show()
```

Interpretation: Week 5 had the highest number of fatality.
